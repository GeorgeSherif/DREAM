```python
import pandas as pd

generated = pd.read_excel("/home/george.ibrahim/Downloads/Semester 2/NLP804/Dream/Generated_2nd.xlsx")
for column in generated.columns:
    generated[column] = generated[column].str.replace("\n", '', regex=False)
generated
```

```python
# Column names excluding 'Prompt' and 'Orig_Text' since we want pairs with 'MSA' and other dialects
dialects = generated.columns[2:]

# Save MSA paired with each other dialect into separate files
for dialect in dialects[1:]:  # Skip 'MSA' itself
    # Select only the 'MSA' and the current dialect columns
    paired_df = generated[['MSA', dialect]]
    paired_df['lang'] = str(dialect)
    # Define a file name based on the dialects
    file_name = f"MSA_and_{dialect}.csv"
    
    # Save the DataFrame to a CSV file
    paired_df.to_csv(f"/home/george.ibrahim/Downloads/Semester 2/NLP804/Dream/Generated/{file_name}", index=False)

# Return a list of saved files' names for confirmation
saved_files = [f"MSA_and_{dialect}.csv" for dialect in dialects[1:]]
saved_files
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Melt the DataFrame to have two columns: one for the text and one for the dialect labels
melted_df = generated.melt(id_vars=['Prompt', 'Orig_Text'], var_name='Dialect', value_name='Text')

# Now, you may want to remove or adjust the 'Prompt' and 'Orig_Text' columns according to your requirements
# For example, to drop these columns and keep just the dialect and its text
final_df = melted_df.drop(['Prompt', 'Orig_Text'], axis=1).rename(columns={'Dialect': 'lang', 'Text': 'sent'})

# If needed, you can reorder the columns or adjust the DataFrame further as required
final_df = final_df[['sent', 'lang']]

final_df.to_csv("/home/george.ibrahim/Downloads/Semester 2/NLP804/Dream/Generated/Generated_test_set_2.csv")
```
